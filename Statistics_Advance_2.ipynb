Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with 
an example.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Probability Mass Function (PMF) for a fair six-sided die
def fair_die_pmf(x):
    if 1 <= x <= 6:
        return 1 / 6
    else:
        return 0

# Probability Density Function (PDF) for standard normal distribution
def standard_normal_pdf(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Plotting PMF for a fair six-sided die
die_outcomes = np.arange(1, 7)
pmf_values = [fair_die_pmf(x) for x in die_outcomes]

plt.bar(die_outcomes, pmf_values, align='center', alpha=0.7)
plt.title('PMF for a Fair Six-Sided Die')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()

# Plotting PDF for standard normal distribution
x_values = np.linspace(-3, 3, 1000)
pdf_values = standard_normal_pdf(x_values)

plt.plot(x_values, pdf_values, label='Standard Normal PDF')
plt.title('PDF for Standard Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

```python
import numpy as np
import matplotlib.pyplot as plt

# Cumulative Density Function (CDF) for a fair six-sided die
def fair_die_cdf(x):
    if x < 1:
        return 0
    elif 1 <= x <= 6:
        return x / 6
    else:
        return 1

# Plotting CDF for a fair six-sided die
die_outcomes = np.arange(1, 8)
cdf_values = [fair_die_cdf(x) for x in die_outcomes]

plt.step(die_outcomes, cdf_values, where='post')
plt.title('CDF for a Fair Six-Sided Die')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.show()
```

Q3: What are some examples of situations where the normal distribution might be used as a model? 
Explain how the parameters of the normal distribution relate to the shape of the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to plot a normal distribution
def plot_normal_distribution(mean, std_dev):
    x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
    pdf_values = norm.pdf(x_values, mean, std_dev)

    plt.plot(x_values, pdf_values, label=f'Mean={mean}, Std Dev={std_dev}')

# Example 1: Standard Normal Distribution (mean=0, std_dev=1)
plot_normal_distribution(0, 1)

# Example 2: Normal Distribution with mean=2 and std_dev=0.5
plot_normal_distribution(2, 0.5)

# Example 3: Normal Distribution with mean=-1 and std_dev=2
plot_normal_distribution(-1, 2)

# Customize plot
plt.title('Normal Distributions with Different Parameters')
plt.xlabel('X')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.show()
```

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli 
Distribution and Binomial Distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom

# Function to plot Bernoulli distribution PMF
def plot_bernoulli_pmf(p, label):
    x_values = [0, 1]
    pmf_values = bernoulli.pmf(x_values, p)

    plt.vlines(x_values, 0, pmf_values, colors='b', lw=5, alpha=0.5, label=label, linestyle='--')

# Function to plot Binomial distribution PMF
def plot_binomial_pmf(n, p, label):
    x_values = np.arange(0, n+1)
    pmf_values = binom.pmf(x_values, n, p)

    plt.vlines(x_values, 0, pmf_values, colors='r', lw=2, alpha=0.7, label=label, linestyle='-')

# Example: Bernoulli Distribution
p_bernoulli = 0.5  # Probability of success
plot_bernoulli_pmf(p_bernoulli, 'Bernoulli Distribution')

# Example: Binomial Distribution (n=1)
n_binomial = 1  # Number of trials for Binomial Distribution
plot_binomial_pmf(n_binomial, p_bernoulli, 'Binomial Distribution (n=1)')

# Customize plot
plt.title('Bernoulli and Binomial Distributions')
plt.xlabel('X')
plt.ylabel('Probability Mass Function (PMF)')
plt.xticks([0, 1])
plt.legend()
plt.show()

```

Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset 
is normally distributed, what is the probability that a randomly selected observation will be greater 
than 60? Use the appropriate formula and show your calculations.

```python
import scipy.stats as stats

# Given parameters
mean = 50
std_dev = 10
observation_value = 60

# Calculate the z-score for the observation value
z_score = (observation_value - mean) / std_dev

# Use the cumulative distribution function (CDF) to find the probability
probability_greater_than_60 = 1 - stats.norm.cdf(z_score)

# Print the result
print(f"The probability that a randomly selected observation will be greater than 60 is: {probability_greater_than_60:.4f}")
```

Q7: Explain uniform Distribution with an example.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to plot the probability density function (PDF) of a uniform distribution
def plot_uniform_distribution(a, b, label):
    x_values = np.linspace(a - 1, b + 1, 1000)
    pdf_values = np.full_like(x_values, 1 / (b - a))

    plt.plot(x_values, pdf_values, label=label)

# Example: Uniform Distribution for a fair six-sided die
lower_bound = 1
upper_bound = 6

# Plot the uniform distribution
plot_uniform_distribution(lower_bound, upper_bound, 'Uniform Distribution (Fair Six-Sided Die)')

# Customize plot
plt.title('Uniform Distribution Example')
plt.xlabel('X')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.show()
```

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data from a non-normal distribution (exponential distribution)
original_data = np.random.exponential(scale=2, size=1000)

# Function to simulate the Central Limit Theorem
def clt_simulation(sample_size, num_samples):
    sample_means = []

    for _ in range(num_samples):
        # Randomly sample data with replacement
        sample = np.random.choice(original_data, size=sample_size, replace=True)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    return sample_means

# Parameters for simulation
sample_size = 30
num_samples = 1000

# Simulate the Central Limit Theorem
sample_means = clt_simulation(sample_size, num_samples)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', label='Sample Means')

# Plot the theoretical normal distribution for comparison
mu_original = np.mean(original_data)
sigma_original = np.std(original_data)
x_values = np.linspace(mu_original - 3 * sigma_original, mu_original + 3 * sigma_original, 1000)
pdf_values = norm.pdf(x_values, mu_original, sigma_original / np.sqrt(sample_size))
plt.plot(x_values, pdf_values, 'r--', label='Theoretical Normal Distribution')

# Customize plot
plt.title('Central Limit Theorem Simulation')
plt.xlabel('Sample Means')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.show()
```
